We need to transform the data so that we remove the `"K"` and then convert the value into a floating numeric value by multiplying by 1000.

To do so we defined a function which would iterate over each entry in a column and use a `if` statement to check if the character "K" was in the entry. If the value contained "K" `.replace` would remove the character and then multiply the stripped value by 1000 to return a floating point number which we could use in our analysis. #### Functions for creating trends for video games 

The code below contains two functions processing_and_preparing_trends and plotting_trends, which were defined to create a graph of trends in game genres over the years.
The processing_and_preparing_trends function is responsible for the initial preparation of the data, it checks if the given columns exist or not. If they are missing, the function catches the error. Then we use the conversion of the release date to the release year using the pd.to_datetime function and changes the Genres column into a list of game genres. The function splits the list into separate rows, which allows for more detailed analysis. Then it checks for missing values ​​again and groups by year and genre, counting the data. Then the function creates a table, where the columns correspond to genres and the rows to years, missing values ​​are filled with zeros. The next function plot_trends uses the previous function to generate the graph and present the trends. It creates axis labels, a legend and a title. It also catches errors related to data availability.

```python
#function to process and prepare data for trends
def processing_and_preparing_trends(df_trends):    
    try:
        #checking whether column exists and process it to extract the year
        if 'Release Date' not in df_trends.columns:
            raise KeyError("The dataset is missing the 'Release Date' column.")
        
        df_trends['Release Year'] = pd.to_datetime(df_trends['Release Date'], errors='coerce').dt.year

        #checking whether column exists and contains valid string values
        if 'Genres' not in df_trends.columns:
            raise KeyError("The dataset is missing the 'Genres' column.")
        
        df_trends['Genres'] = df_trends['Genres'].astype(str)  #converting to strings
        df_trends['Genres'] = df_trends['Genres'].str.strip('[]').str.replace("'", "").str.split(', ')

        #Explode the genres into separate rows and drop missing values(just to make sure, eventhough it was done before)
        dataframe_trends = df_trends.explode('Genres').dropna(subset=['Release Year', 'Genres'])

        #Grouping and counting occurrences
        genre_trends = dataframe_trends.groupby(['Release Year', 'Genres']).size().reset_index(name='Count')

        #Pivot our data for visualization
        genre_trends_pivot = genre_trends.pivot(index='Release Year', columns='Genres', values='Count').fillna(0)
        return genre_trends_pivot
    except KeyError as ke:
        print(f"KeyError: {ke}")
    except Exception as e:
        print(f"An error occurred while processing the data: {e}")
        return None
    
#Function plotting trends
def plotting_trends(pivot_table):
    if pivot_table is None:
        print("Pivot table is none, cannot proceed.")
        return
    try:
        plt.figure(figsize=(14, 8))
        pivot_table.plot(kind='line', figsize=(14, 8), alpha=0.7)
        plt.title("Trends in Gaming Genres Over Time", fontsize=16)
        plt.xlabel("Release Year", fontsize=12)
        plt.ylabel("Number of Games", fontsize=12)
        plt.legend(title="Genres", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.grid(visible=True, linestyle='--', alpha=0.6)
        plt.show()
    except Exception as e:
        print(f"An error occurred while plotting the data: {e}")

```

```python
def plotting_number_of_games(dataset):
    try:
        # Visualize the top teams
        fig, ax = plt.subplots(figsize=(10, 5))
        sns.barplot(
            x = dataset[:10].values,
            y = dataset[:10].index,
            hue = dataset[:10].index,
            palette='muted',
        )
        ax.set_title("Top 10 teams (No. of games released)", fontsize=12, weight='bold')
        ax.set_xlabel("No. of Games", fontsize=8)
        ax.set_ylabel("Team", fontsize=8)

        # Add labels to show the participation of each team
        for i, v in enumerate(team_counts[:10].values):
            ax.text(v, i, str(round(v / len(last_ten_years) * 100, 2)) + ' %', color="black", fontsize=8, weight='bold')

        plt.show()
    except Exception as e:
        print(f"An error occurred while plotting the data: {e}")

plotting_number_of_games(team_counts)
```

Every game developed get some rating from the users community which is averaged out in `"Rating"` column.

The following Function `plot_top_ratings` is trying to visualize and find out top ten teams with highest average rating of all the games developed by them.

```python
def plot_top_ratings(dataset):
    try:
        # Calculate the average ratings for each team
        team_avg_ratings = dataset.groupby('Team')['Rating'].mean().sort_values(ascending=False).head(10)

        # Visualize the average ratings by team
        fig, ax = plt.subplots(figsize=(10, 5))
        sns.barplot(
            x = team_avg_ratings.values,
            y = team_avg_ratings.index,
            hue = team_avg_ratings.index,
            palette='muted'
        )
        ax.set_title("Average Ratings by Team", fontsize=12, weight='bold')
        ax.set_xlabel("Rating", fontsize=8)
        ax.set_ylabel("Team", fontsize=8)

        # Add labels to show the average rating for each team
        for i, v in enumerate(team_avg_ratings[:25].values):
            ax.text(v, i, str(round(v, 2)), color="black", ha="left", va="center", fontsize=8, weight='bold')
        
        plt.show()
    except Exception as e:
        print(f"An error occurred while plotting the data: {e}")

plot_top_ratings(last_ten_years)
```

```python
def calculate_user_distribution(dataset):
    try:
        # Total_players column combines the number of players in these 4 categories (Plays, Playing, Backlogs, Wishlist)
        dataset = dataset.copy()
        dataset['Total_players'] = dataset[['Plays', 'Playing', 'Backlogs', 'Wishlist']].sum(axis=1)

        # Aggregate total players by team and get the top 10 teams
        top_teams = (
            dataset.groupby('Team', as_index=False)['Total_players']
            .mean()
            .sort_values(by='Total_players', ascending=False)
            .head(10)
        )

        # Plot the teams with the highest total players
        fig, ax = plt.subplots(figsize=(10, 5))

        sns.barplot(
            data = top_teams,
            x = 'Total_players',
            y = 'Team',
            hue = 'Team',
            palette = 'muted'
        )

        # Customize the plot
        ax.set_title("Top 5 Teams with the Highest Average Players (Last 10 Years)", fontsize=12, weight='bold')
        ax.set_xlabel("Total Players", fontsize=8)
        ax.set_ylabel("Team", fontsize=8)
        sns.despine(right=True, top=True)

        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"An error occurred while plotting the data: {e}")
```
